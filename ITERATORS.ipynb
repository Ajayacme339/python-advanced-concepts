```python
my_list=[1,2,3,4,5,6]
print(my_list[0])
```

```python
ITERATOR = iter(my_list)
print(ITERATOR)
```

```python
print(type(ITERATOR))
```

```python
print(next(ITERATOR))
```

```python
print(next(ITERATOR))
```

```python
try:
    while True:
        item = next(ITERATOR)
        print(item)
except StopIteration:
    print("Reached the end of the iterator.")
```

```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
```
