### Logging with Multiple Loggers 

```python
import logging
#Create logger 1 

logger_1 = logging.getLogger('module1')
logger_1.setLevel(logging.DEBUG)
#Create logger 2
logger_2 = logging.getLogger('module2')
logger_2.setLevel(logging.WARNING)

#Configure logging settings 
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', datefmt='%Y-%m-%d %H:%M:%S'
                    )
```

```python
# Log messages using different loggers 
logger_1.debug('This is a debug message from module 1')
#logger_1.info('This is an info message from module 1') 
logger_2.warning('This is a warning message from module 2')
#logger_2.error('This is an error message from module 2')
```
