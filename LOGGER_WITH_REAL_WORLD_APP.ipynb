```python
import logging
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s', datefmt='%Y-%m-%d %H:%M:%S')
handelers  = [logging.FileHandler('app1.log'),logging.StreamHandler()]
logger_1 = logging.getLogger("ARITHEMTICAL_OPERATIONS")


def add(a, b):
    logger_1.debug(f"Adding {a} and {b}",{"result": a + b})
    return a + b 

def subtract(a, b):
    logger_1.debug(f"Subtracting {b} from {a}",{"result": a - b})
    return a - b

def multiply(a, b):
    logger_1.debug(f"Multiplying {a} and {b}",{"result": a*b})
    return a * b

def divide(a,b):
    if b == 0:
        logger_1.error("Division by zero is not allowed.")
        return None
    logger_1.debug(f"Dividing {a} by {b}",{"result": a/b})
    return a / b

add(10,5)
subtract(10,5)
multiply(10,5)
divide(10,0)
divide(10,5)



```
